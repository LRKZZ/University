```python
#гистограмма
n_bins = 20
plt.figure(figsize=(10, 6))
plt.hist(data['time_left'], bins = n_bins)
plt.savefig('gistogramma.png')
plt.xlabel('Значения числового признака')
plt.ylabel('Частота')
plt.title('Гистограмма распределения числового признака')
plt.show()
```

```python
# "ящик с усами", или же boxplot
plt.figure(figsize=(10, 6))
plt.boxplot(x=data['ct_score'])
plt.xlabel('Числовой признак')
plt.ylabel('Значения')
plt.title('Диаграмма "ящик с усами" для числового признака')
plt.savefig('boxplot.png')
plt.show()
```

```python
#круговая диаграмма. Для начала подберем подходящий для наглядности столбец.
unique_values_count = data['map'].nunique()
print(unique_values_count)
```

```python
#этот столбец подойдет.
fig, axs = plt.subplots(figsize=(8,8)) # fig будет содержать фигуру, axs - подграфики (в нашем случае один)

count_numbers = data['map'].value_counts() # создаем массив, состоящий из уникальных значений, которые встречаются в этом столбце

axs.pie(count_numbers[1:], autopct='%1.1f%%')
axs.set_title('Круговая диаграмма для номинативного признака')
fig.savefig('krug.png')

plt.show()
```

```python
#тепловая карта
plt.figure(figsize=(10,15))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm', fmt='.2f') # data.corr() - вычисляет корреляцию, annot=True - отображает значения в ячейках, fmt - чтобы после точки отображалось
                                                                 # только 2 цифры.
plt.savefig('teplovaya.png')
```

```python
# диаграмма countplot
f = plt.figure(figsize=(12, 12))
def plot(value):
  if value > 120:
    return "earlygame"
  elif value > 60:
    return "midgame"
  else: return "lategame"

data['stage_of_game'] = data['time_left'].map(plot)
sns.countplot(x = data['is_bomb_planted'], hue=data['stage_of_game'])
plt.savefig('countplot.png')
```

График выглядит так, будто в нем есть ошибка, ибо в midgame и earlygame ни разу не встречалось значение True. Проверим, допущена ли ошибка.

```python
midgame_data = data[data['stage_of_game'] == 'midgame']
true_count = midgame_data['is_bomb_planted'].sum()  # Количество True
false_count = len(midgame_data) - true_count  # Количество False

print(f'В midgame {true_count} раз было True и {false_count} раз было False.')
```

```python
earlygame_data = data[data['stage_of_game'] == 'earlygame']
true_count = earlygame_data['is_bomb_planted'].sum()  # Количество True
false_count = len(earlygame_data) - true_count  # Количество False

print(f'В earlygame {true_count} раз было True и {false_count} раз было False.')
```
