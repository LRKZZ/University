1. Используйте функцию и отрезок ([a, b]) из вашего варианта задания 2
(которые вы использовали для интегрирования). Разбейте отрезок на
столько частей, сколько считаете нужным. Вычислите значение
функции для всех точек получившегося массива.

```python
# Заданный отрезок [a, b]
a = 1.0
b = 56.0

# Количество точек, на которые разбиваем отрезок
num_points = 56

# Разбиваем отрезок на равные части
x_values = np.linspace(a, b, num_points)

# Вычисляем значения функции в каждой точке
def y(x):
    return np.log(np.sqrt(x))

y_values = y(x_values)

# Выводим полученные значения
for x, y in zip(x_values, y_values):
    print(f"x = {x:.2f}, y = {y:.4f}")
```

2. Используя множество точек и множество значений функции в этих

шрифта для названия и подписей. Добавьте на график легенду.

```python
plt.figure(figsize=(8, 6))  # Установим размер графика
plt.plot(x_values, y_values, label='y(x) = ln(√x)')  # Построим график функции и добавим легенду
plt.title('График функции y(x) = ln(√x)', fontsize=12)  # Заголовок графика
plt.xlabel('x')  # Подпись оси x
plt.ylabel('y')  # Подпись оси y
plt.xticks(fontsize=12)  # Размер шрифта для подписей оси x
plt.yticks(fontsize=12)  # Размер шрифта для подписей оси y
plt.legend(fontsize=12)  # Размер шрифта для легенды
plt.grid(True)  # Включим сетку

# Показать график
plt.show()
```

4. Используйте выборки из равномерного и нормального распределения
(из задания 3). Постройте две гистограммы. Используйте различные

```python
# Создание выборок из нормального и равномерного распределений
sample1 = stats.norm.rvs(size=100)
sample2 = stats.uniform.rvs(size=100)

# Построение гистограммы для sample1 (нормальное распределение)
plt.hist(sample1, bins=20, color='blue', alpha=0.7, label='Нормальное')

# Построение гистограммы для sample2 (равномерное распределение)
plt.hist(sample2, bins=20, color='green', alpha=0.7, label='Равномерное')

# Добавление подписей и легенды
plt.title('Гистограммы выборок из нормального и равномерного распределений')
plt.xlabel('Значения')
plt.ylabel('Частота')
plt.legend()

# Отображение гистограмм на графике
plt.show()
```

5. Сгенерируйте выборку из равномерного распределения, состоящую из
50 целых чисел от 1 до 5 (не включая 5). Посчитайте, сколько раз
встречается каждое из чисел в выборке (без циклов!). На основе
полученной информации постройте круговую и столбчатую диаграммы.
Подпишите диаграммы. Для круговой диаграммы укажите подписи
секторов. Для столбчатой диаграммы укажите подписи столбцов. Для
обеих диаграмм укажите множество используемых цветов (для
раскраски секторов и столбцов).

```python
# Генерируем выборку из равномерного распределения
sample = np.random.randint(1, 5, 50)

# Считаем частоту встречаемости каждого числа
unique, counts = np.unique(sample, return_counts=True) # unique - массив из уникальных цифр, counts - массив из количества, как часто встречались эти цифры

# Создадим цвета для диаграмм
colors = ['gold', 'lightcoral', 'lightskyblue', 'lightgreen']

# Построение круговой диаграммы
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.pie(counts, labels=unique, colors=colors, autopct='%1.1f%%', startangle=140) # counts - встречаемость каждого числа, labels - подписывает сектора, autopct - отображение в процентах с 1 цифрой после точки, startangle - начальный угол
plt.axis('equal') # делаем диаграмму кругом
plt.title('Круговая диаграмма')

# Построение столбчатой диаграммы
plt.subplot(1, 2, 2)
plt.bar(unique, counts, color=colors)
plt.xticks(unique, [str(num) for num in unique]) # подписываем столбцы
plt.xlabel('Числа')
plt.ylabel('Частота')
plt.title('Столбчатая диаграмма')

plt.tight_layout()  # Для корректного отображения обоих диаграмм

# Отображение диаграмм
plt.show()
```

6. Используя знания о разбиении отрезка на множество точек (из пункта
1), а также функцию с несколькими переменными из задания 2 (которая
использовалась для решения задачи оптимизации), постройте
трехмерный график функции. Установите цвет линии на графике.
Важно: если в вашем варианте к заданию 2 функция состояла более чем
из двух переменных - используйте только первые две переменные.

```python
def func(x, y):
    return (x + 3)**3 + (y - 3)**2

x = np.linspace(0, 2, 30) # разбиение отрезка
y = np.linspace(0, 2, 30)
X, Y = np.meshgrid(x, y) # двумерная сетка координат
Z = func(X, Y)

fig = plt.figure(figsize=(12, 12))
ax = fig.add_subplot(111, projection='3d')
# 111 - 1 строка, 1 столбец, 1 - индекс подграфика
ax.plot_surface(X, Y, Z, color='skyblue', cmap='viridis')
ax.set_title('Трехмерный график функции')
ax.set_xlabel('Ось X')
ax.set_ylabel('Ось Y')
ax.set_zlabel('Ось Z')
plt.show()
```
