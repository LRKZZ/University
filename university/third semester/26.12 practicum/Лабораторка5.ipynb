9. Найдите билеты по которым плыло 6 и более человек, отобразите список людей (groupby, count).

```python
# создаем df, в котором будут билеты и количество людей за этими билетами
df_task9 = data.groupby("ticket")["name"].count()
# выберем только те билеты, у которых 6 и более людей
tickets = df_task9[df_task9>=6]
tickets
```

```python
# выводим список людей по подходящим условиям билетам
for ticket in tickets.index:
    print("ticket:", ticket)
    passengers = data[data["ticket"] == ticket]["name"]
    print(passengers.to_list())
```

17. Отобразите сколько в каком классе пассажиров выжило/не выжило.

За класс отвечает столбец pclass. Будем сортировать по нему.

```python
# посчитаем количество выживших людей в каждом классе (заодно узнаем количество классов)
df_class = data.groupby('pclass')['survived'].count()
df_class
```

```python
print(f"В первом классе выжило {data[(data['survived']==1) & (data['pclass']==1)].shape[0]} пассажиров")
print(f"В первом классе не выжило {data[(data['survived']==0) & (data['pclass']==1)].shape[0]} пассажиров")
print()
print(f"Во втором классе выжило {data[(data['survived']==1) & (data['pclass']==2)].shape[0]} пассажиров")
print(f"Во втором классе не выжило {data[(data['survived']==0) & (data['pclass']==2)].shape[0]} пассажиров")
print()
print(f"В третьем классе выжило {data[(data['survived']==1) & (data['pclass']==3)].shape[0]} пассажиров")
print(f"В третьем классе не выжило {data[(data['survived']==0) & (data['pclass']==3)].shape[0]} пассажиров")
```
